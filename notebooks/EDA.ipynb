```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
bos = pd.DataFrame(boston.data,columns = boston.feature_names)
bos['price'] = boston.target
```

**Since RAD and TAX are highly correlated, it makes sense to remove one of those two features**

```python
Q1_list = []
Q3_list = []
IQR_list = []

for feature in bos.columns.drop(['CHAS','RAD','price']):
    Q1 = np.round(np.percentile(bos[feature],25),2)
    Q1_list.append(Q1)
    Q3 = np.round(np.percentile(bos[feature],75),2)
    Q3_list.append(Q3)
    IQR = np.round(Q3 - Q1,2)
    IQR_list.append(IQR)

print(f"Q1: {Q1_list}")
print()
print(f"Q3: {Q3_list}")
print()
print(f"IQR: {IQR_list}")
```

```python
## removing outliers by replacing them with the null values
## here we will exclude features namely CHAS, RAD since they are kind of categorical and price since it is a target column
lst = bos.columns.drop(['CHAS','RAD','price'])
for i in range(len(lst)):
    upper_limit = Q3_list[i] + 1.5*IQR_list[i]
    lower_limit = Q1_list[i] - 1.5*IQR_list[i]

    bos[lst[i]] = np.where((bos[lst[i]] > upper_limit) | (bos[lst[i]] < lower_limit), np.nan, bos[lst[i]])
```

```python
bos.isnull().sum()
```

```python
## now we will fill the null values using the mean of the values from the column
for feature in bos.columns:
    bos[feature].fillna(np.mean(bos[feature]),inplace=True)
```

```python
## checking the null values in the columns again
bos.isnull().sum()
```

#### Preprocessing operations to be done
1. Removing one of the features from RAD or TAX
2. replacing outliers with the null values
3. replacing null values with the mean of the column

```python
## separating dependent and independent features
X = bos.drop(columns=['price'])
y = bos['price']
```

```python
## creating a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lr = LinearRegression()
score_list = cross_val_score(lr,X,y,cv=5)
print(f"Cross validation score: {np.round(score_list.mean(),2)} +/- {np.round(score_list.std(),2)}.")
```
